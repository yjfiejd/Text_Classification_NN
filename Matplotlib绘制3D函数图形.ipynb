```python
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
fig1 = plt.figure()
ax = Axes3D(fig1)
plt.show()
```

```python
import numpy as np
X = np.arange(-2, 2, 1)
Y = np.arange(-2, 2, 1)
print(X)
print(Y)
print("------------")
X,Y = np.meshgrid(X, Y)
print(X)
print(Y)
```

```python
def fun(x, y):
    return np.power(x, 2) + np.power(y, 2)

fig1 = plt.figure()
ax = Axes3D(fig1)
X = np.arange(-2, 2, 0.1)
Y = np.arange(-2, 2, 0.1)
X, Y = np.meshgrid(X, Y)
Z = fun(X, Y)
plt.title("Title")
#用取样点（X, Y, Z）构造曲面
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap = plt.cm.coolwarm)
ax.set_xlabel('x label', color = 'r')
ax.set_ylabel('y label', color = 'g')
ax.set_zlabel('z label', color = 'b')
plt.show()
```
